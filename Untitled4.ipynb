```python
import geopandas as gpd

# Yüklenen dosyayı oku
gdf = gpd.read_file("istanbul-admin-level-8.geojson")

# İlk birkaç satıra göz at
gdf.head()
```

```python
# Tüm geometrileri birleştir (union) ve merkezini bul
istanbul_union = gdf.unary_union
center_point = istanbul_union.centroid

print("İstanbul'un coğrafi merkezi (tahmini):", center_point)
```

```python
import matplotlib.pyplot as plt

# Haritayı çiz
gdf.plot(edgecolor='black', facecolor='none', figsize=(10, 8))

# Merkezi işaretle
plt.plot(center_point.x, center_point.y, 'ro', markersize=10)
plt.text(center_point.x + 0.01, center_point.y, "Merkez", color='red')

plt.title("İstanbul Haritası ve Coğrafi Merkezi")
plt.xlabel("Boylam")
plt.ylabel("Enlem")
plt.show()
```

```python
import geopandas as gpd
from shapely.geometry import Point

# Coğrafi merkez noktası
center_point = Point(28.80784047388923, 41.15795996884021)  # Bu değişkenler daha önce hesaplanmıştı

# Merkez noktayı içeren ilçeyi bul
district_containing_center = gdf[gdf.contains(center_point)]

if not district_containing_center.empty:
    ilce_adi = district_containing_center.iloc[0]['name']  # veya 'name:tr' olabilir
    print(f"Coğrafi merkez {ilce_adi} ilçesi sınırları içerisinde yer alıyor.")
else:
    print("Coğrafi merkez hiçbir ilçeyle kesişmiyor.")
```
